*What is an agent?*
It can be thought of as an interface with a variety of tools chained together. First do this, then do that ...

These tools are defined by users as prompts to the LLM. Connecting an agent with a LLM simply brings these tools into the arsenal of the LLM, enabling it to use them.
The role of the user is to provide the LLM with the tools through prompts, while the LLM is responsible for deciding which tools it must use depending on the user input.

Another thing to observe is that there can be as many agents as tools that you might create. So, an agent could search the web, could fetch weather data, could perform calculations, or even do Machine Learning.

## **Hypothesis Testing**

In Data Science and Statistics in general, the concept of **Hypothesis Testing is a prominent part of the method of operations.**

Hypothesis Testing is just a fancy term for **statistically fact-checking a quantitative claim.**

We pose a particular claim (an alternate hypothesis) and we pose a null claim (a claim which says that our hypothesis is useless). After this we take several examples and see whether they follow our claim or not with a degree of assurance so that those results could not have occurred by chance.

In the example below, we follow the following structure:

**Claim:** Obesity increases the risk of heart disease

Now we would use the following steps:
- On average, what is the obsesity level of people with heart disease and without heart disease ?
- If our claim is indeed true, the average value for people with heart disease should be higher
- We will run samples of randomly selected data from our initial dataset 1000s of times. We will then see what the distribution of the averages looks like.
- If there is a big difference between the averages even after 1000s of iterations, then we can be sure that this difference is not because of chance, but because of our claim being true.

In the following code block, we will be creating a toy SQL database using SQLite for demonstration purposes.

## **Retrieval-Augmented Generation**

**In this section, we will perform Retrieval-Augmented Generation (RAG) with a LangChain agent.** As we know, RAG refers to the process of supplementing the context of an LLM from the System Prompt and User Query, with additional chunks of text that may have been retrieved from external sources such as a Document or the results of a Web Search. This additional context is used by the LLM to generate what usually turns out to be a much more contextual and relevant answer to the User Query, especially when that query refers to current news or specific facts or figures that may not have been present in the LLM's original pre-training dataset. **The benefits of augmenting an LLM's capability with RAG are numerous**, such as providing it with additional knowledge, reducing its risk of hallucination, and being able to now fact-check the LLM by monitoring the sources from which it has generated its answer.

- What we shall do here is, we shall first load our document, split it into easy to handle chunks, and then vectorize and store these chunks in a Vector Database
- We provide our RAG Agent with this Vector Database and a few more tools
- Whenever we ask the agent a question, it will refer to the document to find something from the database which could answer our question
- So if you asked it to summarise a chapter from the book, the agent would first fetch the specific chapter from the database and then summarise it with the LLM

Let's now see the agent in action.

```python
from langchain.agents.agent_toolkits import create_retriever_tool
```

We shall import a chapter from an Operating Systems course book.

```python
# Let's create an object to load the book
Loader = PyPDFLoader('https://pages.cs.wisc.edu/~remzi/OSTEP/cpu-intro.pdf')
```

Next, we shall load the document and split it into several chunks, each of size 1000 and with an overlap of 50 between chunks.

```python
# We use the loader created above to load the document
documents = Loader.load()

# We split the document into several chunks as mentioned above
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=50)
texts = text_splitter.split_documents(documents)
```

Let's vectorize these chunks into vector embeddings, and store them in the FAISS Vector Database.

```python
# Now we transform these chunks into the list of number called vectors and store them into our database
embeddings = OpenAIEmbeddings()
db = FAISS.from_documents(texts, embeddings)
```

Now, we create a Retriever instance that will perform similarity search and extract those vector chunks relevant to the user query.

```python
# We create a retriever instance. Its job is to fetch for us a vector (portion of the document) relevant to the user
# question
retriever = db.as_retriever()
```

Finally, we shall write the prompt to create the RAG agent.

```python
# This is the prompt to create a RAG agent for us
retriever_name = "search_the_text_of_pdf"
retriever_desc = """The purpose of this tool is to answer questions based on the book: 'Operating Systems 3 \
easy pieces' regarding operating systems in computer science. Answer any query input by the user from the vectorbase\
if you do not know the answer, then use the serpapi tool to search the web for the answer. Clearly state that you\
searched the web. Keep your answers short and precise."""
```

We are defining two tools below:

1. **RAG Tool:** This is the tool that performs RAG for us on the loaded document. It needs access to the retreiver object that we created above to fetch relevant pieces from our database to answer our questions.

2. **Serpapi Tool:** In case the LLM is unable to find an answer for some question from the document that we uploaded, one option is to tell you that it did not find any relevant answer from the document, so it would need to search the internet to give you an answer. The Serpapi tool searches Google for the result.

```python
rag_tool = create_retriever_tool(
    retriever,
    retriever_name,
    retriever_desc
)

# My free tier SERPAPI API key
os.environ['SERPAPI_API_KEY'] = 'd6f4685c4ebac6d2728a5331829fd855f93b88859254abf7e0498dd32c34da59'
#os.environ['SERPAPI_API_KEY'] = '4bde37c74068633153b825cb3ff392ee3a6697a630674d5abb5d55be50f58a49'
search_tool = load_tools(['serpapi'])
tools = [rag_tool, search_tool[0]] # we can have multiple tools, hence the list
```

### Conclusions ###

#### TL;DR
* For AI to be useful, you must be able to take a step beyond simply answering questions, or drawing a conclusion. You must be able to take action based on your well-reasoned and transparent conclusion. The agent framework provides a mechanism to drive actions. While an LLM can provide concise summarization and clear answers, adding the agent framework provides real utility.
* Agents incorporate domain knowledge in their design, and are an excellent mechanism to leverage and distribute the domain expertise of the author.
* Both identifying or developing agents to use requires a solid grounding in the subject matter.
* There's still no "AI magic" that enable agents to replace skilled human data analysts. Agent outputs still require critical analysis, but the effort to produce them can be streamlined.

#### Mechanics
* Retriever assistants are a powerful mechanism to utilize general chat LLMs to leverage curated and well maintained proprietary data - that data is the unique business differentiation that can produce a useful product.
* The agents seem to persist some memory regarding the shape and nature of the data, but they are mostly flexible enough to deal with changes of the data as it flows through the workflow. As columns in the data set are transformed, agents seem to be able to recognize those transformations and adapt.
* For most tasks, this is not a factor as data is analyzed through a well-designed workflow.
* However, it would be trivial to copy the data, and add an instruction to reload the original data for each agent operation.
* There's some variation of the outcomes for each execution of the agent. I attempted to minimize that by explicitly setting temperature to 0, however I did still see variability in the results.

#### Error Handling
* It's poor. Every agent incorporates some level of error handling, but the results are inconsistent and occasionally fail.
* The agent initialization option to handle parsing errors also works inconsistently.
* Re-running the same instructions often succeeds when the initial execution failed over syntax, parsing, or data frame errors the agent didn't handle 

#### Environmental Considerations
* As is the case with all other aspects of the current state of the art in AI, the agent libraries, their dependencies, and the operational environment is incredibly fragile and unstable. The skillset required to maintain it is significant (read:expensive)
* If you're relying on this technology to get real work done in your business, you'll need a stable and consistent ecosystem. Fragile environments scale poorly, and the impact of outages increases with the value the solution produces.
* It's easy to spend more effort configuring, troubleshooting, and re-configuring the environment than on actually completing the analysis.
* It's common to completely reconfigure the environment and dependencies from within a workflow. That's poor engineering practice, and will inevitably produce inconsistent outcomes.
* The reliance on a narrowly defined working environment for each task significantly complicates moving from prototype to production, and makes operational change control virtually impossible.
* The requirement to downgrade some libraries for stability means compromising with dependent libraries to exclude bug fixes and new features.
* The ecosystem of distributed development - frequently by authors lacking software engineering skills - means much higher costs for support than even most typical FOSS solutions.
* The requirement to highly customize virtual environments for each task - in order to control dependencies - adds unecessary operational costs.

#### Data Security
* This raised questions for me of where my data would be visible, and what leaves my environment. How much of my data is sent through the agent, and how can I demonstrate policy compliance to my customers? The agent framework makes securing proprietary data extremely challenging.
* Also - understand where these processes are executing, and how costs are allocated with an external service can complicate pricing and budgeting decisions.
* There's certainly a lack of transparency for data in motion - I think that arises from the notion that the data available to most LLMs are public, but raises important questions where retrieval agents and proprietary data are used.
* The leakage of personal or company proprietary data can result in liabilities that can destroy a company, and damage entire industries. Security is not optional.
* Agents complicate the evaluation of bias or explanability by obscuring transparency. You'll have a difficult time demonstrating to customer that your solution is balanced and fair.

### Agent Use Cases

#### Bug auto-responder
* One interesting use case I'm seeing implemented already is an immediate auto-responder for bug submissions. A bug report is structured data, and a responder can be tuned for specific domains. The auto-responder can offer a quick search of related bug reports, a pointer to duplicate cases, an analysis of the stack trace, troubleshooting or additional data collection steps that can be followed to reduce the time required for a human to get back to the submitter. They basically save everyone some time, and shorten time to resolution. One key here is setting expectations appropriately - rather than try to pretend that the response was thoughtfully generated by a human subject matter expert, the auto responder sets the expectation that they are providing quick "first responder" level assistance. Reviewing each bug as it arrives can also help the support team triage and prioritize, and route bugs to the appropriate dev team.

#### Operational help desk assistant
* Somewhat related is an operational troubleshooting assistant that can consume alarms from monitoring systems, and trouble ticket reports to cluster and relate them into incidents. A troubleshooting assistant can recommend steps (or take actions) to test the environment, or recall root causes of past incidents with a similar chain of events. An operational assistant can help standardize troubleshooting or recovery processes, and provide some consistency for a distributed help desk team. Like bug reports, they can help triage and route tickets to the correct team or teams to reduce time to repair. There are commercial solutions that have been available for over a decade in this space.

#### Writing assistant
* While there's certainly significant value in concise summation, grammar or spelling correction, and suggestions to organize text more clearly, the real value of a writing assistant comes with an LLM that's tuned for the language and concepts of a specific domain. Coupled with a retriever that can reference internal documentation and curated data, you can produce a result that gives a common "voice" to company communications. For companies that are innovating, there's always a need for new original writing to drive the key messaging and differentiation for the product. But these tools can be a powerful accelerator for staff writers producing both internal and external communications.

#### Graphic design assistant
* Along the same lines, AI based tools can accelerate the creative ideas of your graphic designers to produce unique content much faster. Like the writing assistant, you can use AI to drive common "look and feel" in your graphics. Doesn't take the place of original concepts, but it can certainly shorten the time to get new designs out.

#### Business analyst assistant
* In the corporate ecosystem, I think of data scientists developing and testing algorithms, software engineers building the toolset and ecosystem, and business analysts with customer and domain knowledge using those tools to make well grounded and supported business decisions. I see many companies trying to cast business experts into the role of data scientists, or data scientists into the role of software engineers. I think these are distinct competencies, and smart companies form ensemble teams to work together and leverage the strengths of various roles. These tools can be powerful in the hands of a business analyst, but only if they are reliable and easy to use, and the outcomes are grounded in sound and comprehensive analysis. The best tools are tuned for the domain, incorporate the terminology and concepts of the market, and retrieve relevant proprietary company data to produce insights.

#### Superior search
* The current search paradigm - web crawling, keyword matching, and endless pages of links - is choking to death on techniques to game the results, like SEO and corrupted reputation systems. It's rapidly becoming unusable. It's always required what I like to call "gray-matter correlation"; that is, you perform the last step of determining relevance of the results by scanning and incorporating your own experience and expertise. But it's worse now as the markers for reputable information become more obscured. Up until now you've needed "google-fu", some expertise to determine keywords and search criteria to get the result you want. With an LLM, you are able to get more more nuanced and detailed conclusion, but you need a new skillset: prompt engineering. So - the promise here for researchers is a concise and complete conclusion, but you need to learn to ask for it. Again - coupling an LLM with company and domain specific data available through a retriever offers the promise of a better quality experience to find answers for your customers. 

### Business Recommendations
#### Kids: don't try this at home
Given the current rapidly evolving state of ML/AI, for nearly all companies this will be a __buy__, not a __build__ play. The cost of building infrastructure to train your own models, to build and maintain an ML pipeline, hire the talent required in an extremely competitive market, train analysts, etc. is prohibitive. You simply won't see the value. 

You should look for a delivery partner that can maintain a stable infrastructure, offer significant ease of use, and provide predictable transparent pricing. Vet your use cases, and match them against an external service provider catalog to make a clear decision about whether you'll actually produce some value. Consider carefully your opportunity cost when you assess your time to market.

Use a provider with a mature toolkit. No one wants to pin their business on an AI that "hallucinates", or becomes "lazy."

The state of the art now relies upon fragile software, powerful infrastructure, and very specialized skills to showcase some spectacular capabilities. These are far from economical and reliable business services. While it's very useful to understand what's required to deliver, don't be tempted to try to build out model training or data science basic research in house. Be a smart consumer.

#### "Nobody buys math" (j.reves)
The underlying math and analytic techniques for machine learning are free commodities. Academia will continue to research more powerful and efficient algorithms that will find their way into the math libraries and models that ML depends upon. No one will write you a check for your math.

The valuable commodity is data. What you have, that no one else has.

Customers will pay for the kinds of problems you can solve for them that no one else can solve. You do that with data - raw data, or engineered features - that are unique among your competitors.

The initial wave of LLMs depended largely on "free" text and data exposed on the internet, and scraped to be used to train these models without permission or attribution. The controversy and legal challenges will linger, but going forward the really valuable solutions will depend on well-curated, unique sets of data.  We're already seeing vendors whose engineering staff had previously allowed unrestricted access shut down or rate limit programmatic access. We're going to see a change in the notion, "Data wants to be free" as companies realize what they know is precious. Data does __not__ want to be free.

#### Focus on domain expertise
You are better off teaching your domain experts what is possibly with ML, and keeping them focused on problems in their market space. Use the tools to solve problems; don't go down the rathole of building out your company just to build ML tools.

Hire the IT staff you need to make effective use of those tools. Hire the data science team you need to understand how to apply them. You'll see greater value from subject matter experts in your domain than from technicians and mathematicians. Keep track of the tools and analytic capabilities, for sure, but be mindful of taking on the cost of in-house development when an entire industry is focused on doing your development for you.

No one does your business better than you do. Everyone else is busy developing new tools you can use.  Let them.

#### Positioning LLMs
Position LLMs and generative solutions as productivity assistants, and encourage your staff to use them thoughtfully and carefully. These tools are not replacing high-value strategic analysis at this point.  They can help get your existing staff be productive, and faster. You can spend less time on repetitive or tedious tasks, and focus on the important stuff.

#### Product Management is critically important
Your product management team has a new stakeholder: in addition to engineering, user experience (design and research), marketing and communications, and operations - they are working with dedicated data science teams. Data science has a unique perspective on the world that's not necessarily represented by any of these other groups. They play a critical role in data management in collecting / cleaning / engineering your customer and product data. They can quickly and objectively help to validate a hypothesis, and identify insights that are actionable for the business.

They are not, however, software engineers. 

Don't make the mistake of assuming that since data scientists write code, they are software engineers. They are not. They are mathematicians, with a fundamentally different world view than your engineering or operations teams. You cannot expect to take exploratory code from your data science team, and implement it directly in production. You need the engineering and operations perspective to build a robust service.

For Product Managers, this means you need to learn to speak a new language.  You might think initially that you can use the same vocabulary and conceptual framework to speak with data scientists that you use with software engineers.  You can't. It's a new set of terminology, a new mindset and you need to be prepared to translate.
